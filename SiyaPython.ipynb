```python
#function 1.new list with unique elements
def unique_list(numbers):
    unique=[]
    for item in numbers:
        if item not in unique:
            unique.append(item)
    return unique
print("unique elements::",unique_list([10,23,32,44,10,32,42,22]))
```

```python
#functions 2.prime
def prime(a):
    for i in range (2,a) :
        if a % i == 0 :
            return "This is not prime number"
    return "Prime number"
a = int(input("Enter a number : "))
print(prime(a))
```

```python
#function 3.perfect number
def perfect_number(n):
    sum = 0
    for x in range(1, n):
        if n % x == 0:
            sum += x
    if(sum==n):
        print(n)
    return "Number is Perfect number"
n=int(input("Enter the number to check:"))
print(perfect_number(n))
```
